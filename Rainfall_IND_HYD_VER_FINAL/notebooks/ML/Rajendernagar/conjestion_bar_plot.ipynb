```python
import config
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path_of_the_directory= config.DATA+'//data//ml_results_mosaic//Grab_&_Go'
#print("Files and directories in a specified path:")
metrics_dict = {}
for filename in os.listdir(path_of_the_directory):
    f = os.path.join(path_of_the_directory,filename)
    if os.path.isfile(f):
        if f.endswith(".csv"):
            # print(f)
            # print(filename.replace('_model_metric.csv', ""))
            metrics_dict[filename.replace('_model_metric.csv', "")]=pd.read_csv(f)
```

```python
model_df = pd.DataFrame(columns=['Name','Metric'])
for key in metrics_dict:
    # print(metrics_dict[key]['Metric'].max())
    # ax = plt.figure(figsize=(10, 10)
    model_df.loc[len(model_df)] = key,metrics_dict[key]['Metric'].max()
    #plt.bar(x = key, height= metrics_dict[key]['Metric'].max())
model_df.sort_values(ascending=False, by = 'Metric').plot.bar(x = 'Name', y = 'Metric', title = 'Mosaic Bay - Grab & Go')
plt.ylim(0,1)
plt.ylabel('$R^2$ Accuracy')
plt.xlabel('Item name')
plt.ylim(0,1)
```
